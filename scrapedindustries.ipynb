```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns

%matplotlib inline

import requests
from bs4 import BeautifulSoup
```

```python
industry_state = 'https://www.bls.gov'
industry_url = f'{industry_state}/oes/current/area_emp_chart/area_emp_chart_data.htm#Alabama'

print(industry_url)
```

```python
resp = requests.get(industry_url)

print(resp.status_code)
```

```python
soup = BeautifulSoup(resp.text, 'html.parser')
```

```python
tables=soup.findAll ('table')
```

```python
title=tables[1].find ('caption').text
```

```python
table=pd.read_html(str( tables[1]))[0]
```

```python
##find date and name it
table['title']=title
print(table)
```

```python
##Scrape all the data from page. 50 states plus top 10 occupations
dataframes = []
for souptable in tables[1:]:
    title=souptable.find ('caption').text
    dataframe=pd.read_html(str( souptable))[0]
    dataframe['title']=title
    print (dataframe.shape)
    ##result = t.text.strip()
    dataframes.append(dataframe)
```

```python
final_dataframe=pd.concat(dataframes, ignore_index=True)
```

```python
final_dataframe.head(551)
```

```python
final_df=final_dataframe[:550]
```

```python
f_df=final_df['title'].str.extract(r'Largest occupations in (.*),')
```

```python
D_F = pd.concat([f_df, final_df], axis=1)
```
