```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns

%matplotlib inline

import requests
from bs4 import BeautifulSoup
```

```python
gdp=pd.read_csv ('per capita gdp by state 050718(1).csv', header=4)
gdp
```

```python
##Clean Nan data
gdp1=gdp.drop ('Unnamed: 15', axis=1)
```

```python
#Get the mean of select data
col=gdp1.loc[:, "2008":"2017"]
gdp1['gdp_mean']=col.mean(axis=1)
gdp1
```

```python
##Rename data to make merge simple
gdp1.rename(columns={'Area': 'State'}, inplace=True)
gdp1.head()
```

```python
gdp1.to_csv('GDP_PerCapita.csv')
```

```python
##Read in second data set
df2=pd.read_csv ('school_perstudent.csv')
df2
```

```python
##Get rid of uneeded data
School=df2.drop(df2.loc[:,'Unnamed: 25':'Unnamed: 50'].head(0).columns, axis=1)
School
```

```python
## reduce dataset I am working with on Tableau
school_per_student=School.drop(School.loc[:, '1993': '2007'].head(0).columns, axis=1)
school_per_student
```

```python
##third dataset
pop=pd.read_csv ('total students per state.csv')
pop
```

```python
School.to_csv('school-expense-per-student_full.csv')
```

```python
merged_gdp_school=pd.merge(gdp1, school_per_student , on='State')
```

```python
##Merge all datasets together
merge_total=pd.merge(merged_gdp_school, pop, on='State')
merge_total
```

```python
merge_total.to_csv('merge_total.csv')
```
